```python
import lxml.etree
from nltk.tokenize import sent_tokenize
import re
from nltk.parse.stanford import StanfordDependencyParser
path_to_jar = 'resources/parser/stanford-corenlp-3.9.1.jar'
path_to_models_jar = 'resources/parser/stanford-corenlp-3.9.1-models.jar'
dependency_parser = StanfordDependencyParser(path_to_jar=path_to_jar, path_to_models_jar=path_to_models_jar)
```

```python
with open('datasets/Biomed/aimed_bioc.xml', 'r') as myfile:
    data=myfile.read().replace('\n', '').replace('.', '. ')
```

```python
root = lxml.etree.fromstring(data)
documents = root.findall('document')
```

```python
def protein_extraction(document):
    annotations = document.findall('passage/annotation')
    protein_dict = {}
    for annotation in annotations:
        protein_dict[annotation.attrib['id']] = annotation.find('text').text
    return protein_dict
```

```python
def relation_extraction(document):
    relations = document.findall('passage/relation')
    relation_pairs = []
    for relation in relations:
        nodes = relation.findall('node')
        relation_pairs.append((nodes[0].attrib['refid'], nodes[1].attrib['refid']))
    return relation_pairs      
```

```python
def extract_root(sentence):
    result = dependency_parser.raw_parse(sentence)
    dep = result.__next__()
    return list(dep.triples())[0][0][0]
```

```python
for document in documents:
    proteins = protein_extraction(document)
    relations = relation_extraction(document)
    sentences = sent_tokenize(document.find('passage/text').text)
    
    filtered_sentences = []
    for sentence in sentences:
        for relation in relations:
            if proteins[relation[0]] in sentence and proteins[relation[1]] in sentence:
                filtered_sentences.append(sentence)
                
                
    for sentence in filtered_sentences:
        print(sentence)
        print(extract_root(sentence))
    print('-----')
```
